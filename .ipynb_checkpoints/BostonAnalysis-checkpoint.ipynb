```python
# detect and remove outliers using z-score
from scipy import stats

z_scores = stats.zscore(df)
abs_z_scores = abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df = df[filtered_entries]
print('Number of instances after removing outliers:', df.shape[0])
```

```python
# check for missing values
print('Columns with missing values:', df.columns[df.isnull().any()].tolist())
```

```python

# create a copy of the original dataframe
df_copy = df.copy()

# replace outliers with the median for each feature
for col in df_copy:
    q1 = df_copy[col].quantile(0.25)
    q3 = df_copy[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df_copy.loc[df_copy[col] < lower_bound, col] = df_copy[col].median()
    df_copy.loc[df_copy[col] > upper_bound, col] = df_copy[col].median()
```
